```python
# here i load the file and print the first rows
# this helps me quickly see the columns and values
def load_data(path):
    df = pd.read_csv("/content/sample_data/california_housing_train.csv")
    # remove extra spaces from column names
    df.columns = df.columns.str.strip()
    print("first 5 rows",df.head())
    return df
```

```python
# this function prints number of rows and columns
# this helps me understand the size of the dataset
def check_shape(df):
    print("dataset shape")
    print("rows:", df.shape[0])
    print("columns:", df.shape[1])
    return df
```

```python
# here i use info() to check data types and missing values
# this helps me see if some columns have wrong types
def inf_types(df):
    print("data types info",df.info())
    return df
```

```python
# here i check numeric columns even if they already look correct
# sometimes datasets read numbers as text so i make sure they are numeric
def fix_type(df):
    df["total_rooms"] = pd.to_numeric(df["total_rooms"], errors="coerce")
    df["total_bedrooms"] = pd.to_numeric(df["total_bedrooms"], errors="coerce")
    print("checked numeric types for rooms columns")
    return df
```

```python
# here i count missing values in every column
# this helps me know which columns need cleaning
def missing_values(df):
    print("missing values per column")
    print(df.isnull().sum())
    return df
```

```python
# here i fill missing values based on the column type
# lot frontage is a numeric column
# i use median because it is better when data has outliers because it is robust to outliers and skewed data
# median is more stable than mean
def handle_missing(df):
    df["total_bedrooms"] = df["total_bedrooms"].fillna(df["total_bedrooms"].median())
    print("handled missing values in total_bedrooms")
    return df
```

```python
# here i check if there are duplicate rows
# duplicates can affect analysis so i remove them
def handle_duplicates(df):
    print("duplicate rows:", df.duplicated().sum())
    df = df.drop_duplicates()
    print("shape after removing duplicates:", df.shape)
    return df
```

```python
# here i check outliers in the target column saleprice
# i also calculate the iqr

def handle_outliers(df):
    # boxplot to see outliers visually
    sns.boxplot(x=df["median_house_value"])
    plt.title("boxplot of  median house value")
    plt.show()
    # calculate q1 and q3
    q1 = df["median_house_value"].quantile(0.25)
    q3 = df["median_house_value"].quantile(0.75)
    # calculate iqr
    iqr = q3 - q1
    print("iqr value:", iqr)
    return df
```

```python
# final checks to confirm data looks correct
def final_checks(df):
    print("nullsss in target:", df["median_house_value"].isnull().sum())
    print("all target values > 0:", (df["median_house_value"] > 0).all())
    print("total columns:", df.shape[1])
    return df
```

```python
# this function runs all cleaning steps in order
def cleaning_pipeline(path):
    df = load_data(path)
    df = check_shape(df)
    df = inf_types(df)
    df = fix_type(df)
    df = missing_values(df)
    df = handle_missing(df)
    df = handle_duplicates(df)
    df = handle_outliers(df)
    df = final_checks(df)
    print("cleaning done")
    return df
```

```python
clean_df = cleaning_pipeline("/content/AmesHousing.csv")
```
